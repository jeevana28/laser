```python
import os
import pickle
from pprint import pprint

# Specify the directory containing the pickle files
directory = '/Users/jeevana/Documents/GitHub/laser/coco/results/blip_results/BLIP/rank-reduction/fc_in/'

# Loop over each file in the specified directory
for filename in os.listdir(directory):
    # Check if the file is a pickle file (you can modify the extension as needed)
    if "evaluation" in filename:
        file_path = os.path.join(directory, filename)
        
        # Load the pickle file
        with open(file_path, 'rb') as file:
            loaded_data = pickle.load(file)
        
        # Print the filename and its contents
        print(f"Contents of {filename}:")
        pprint(loaded_data)
        print("\n" + "="*50 + "\n")  # Separator between files for readability
```

```python
original_size = os.path.getsize("original_model.pth")
reduced_size = os.path.getsize("reduced_model.pth")

print(f"Original model size: {original_size / 1e6:.2f} MB")
print(f"Reduced model size: {reduced_size / 1e6:.2f} MB")
```

```python
import torch
# Load models from saved paths
# Load the models directly
original_model = torch.load("original_model.pth")
reduced_model = torch.load("reduced_model.pth")

# Access weights of a specific layer
original_weights = original_model.layer.weight.data
reduced_weights = reduced_model.layer.weight.data

# Compare weights
diff = torch.abs(original_weights - reduced_weights)
mean_diff = torch.mean(diff).item()
max_diff = torch.max(diff).item()

print(f"Mean difference: {mean_diff:.6f}, Max difference: {max_diff:.6f}")
```

```python
from transformers import BlipForConditionalGeneration, BlipProcessor
from torchviz import make_dot
import torch

# Load the BLIP model
model_name = "Salesforce/blip-image-captioning-large"
model = BlipForConditionalGeneration.from_pretrained(model_name)

# Create a dummy input
processor = BlipProcessor.from_pretrained(model_name)
dummy_input = processor(images=torch.zeros((1, 3, 224, 224)), return_tensors="pt")

# Forward pass through the model
outputs = model(**dummy_input)

# Visualize the computational graph
graph = make_dot(outputs.logits, params=dict(model.named_parameters()))
graph.render("blip_model_architecture", format="png")
```
